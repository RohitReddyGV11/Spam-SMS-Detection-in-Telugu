```python
import matplotlib.pyplot as plt
import pandas as pd

# Model names and F1 Scores
data = {
    "Model": ["IndicBERT", "CNN", "mBERT", "MuRIL", "Logistic Regression", "Random Forest", "Naive Bayes"],
    "F1 Score": [1.0000, 0.9988, 0.9988, 0.9894, 0.9748, 0.9736, 0.9055]
}

# Create DataFrame and sort
df = pd.DataFrame(data).sort_values(by="F1 Score", ascending=False)

# Color palette similar to your reference image
colors = [
    "#4C6EF5",  # deep blue (CNN-like)
    "#5C7AEA",  # light blue
    "#6D8DF3",  # medium blue
    "#A3B8F0",  # pale blue
    "#D3D8E2",  # greyish tone
    "#EAA39A",  # soft red
    "#D77A61"   # warm red (lowest)
]

# Plot
plt.figure(figsize=(10, 6))
bars = plt.barh(df["Model"], df["F1 Score"], color=colors, edgecolor="none")

# Add F1 score labels beside each bar
for bar in bars:
    width = bar.get_width()
    plt.text(width - 0.08, bar.get_y() + bar.get_height()/2,
             f"{width:.4f}", ha="center", va="center", color="white", fontsize=10, fontweight="bold")

# Labels and title
plt.xlabel("F1 Score", fontsize=12)
plt.ylabel("Model", fontsize=12)
plt.title("Final ML + DL Model Comparison by F1 Score", fontsize=14, fontweight="bold")

# Reverse order to keep highest at top
plt.gca().invert_yaxis()

# Subtle grid and layout tweaks
plt.grid(axis='x', linestyle='--', alpha=0.3)
plt.tight_layout()

# Show plot
plt.show()
```
